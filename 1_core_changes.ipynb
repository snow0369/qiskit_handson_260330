# 2. From `execute()` to Primitives

In early Qiskit versions, circuit execution commonly used:

```
execute(circuit, backend)
```

Qiskit 2.x introduces **primitives** as the standard execution interface.

Two key primitives:

- **Sampler** → sampling measurement results
- **Estimator** → expectation values of observables


### Example: Using a Primitive Sampler

````python
from qiskit.circuit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.measure_all()

sampler = StatevectorSampler()
job = sampler.run([qc])

result = job.result()
result
````

# 3. Improved Transpiler Workflow

The **transpiler** converts abstract quantum circuits into hardware-compatible circuits.

Qiskit 2.x improves:

- hardware-aware compilation
- pass manager usability
- modular compilation pipelines

Instead of manually building pipelines, users can use **preset pass managers**.


````python
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

pm = generate_preset_pass_manager(optimization_level=3)
optimized_circuit = pm.run(qc)

optimized_circuit
````

# 4. Parameterized Circuits

Parameterized circuits are important for hybrid algorithms such as:

- VQE
- QAOA
- Quantum machine learning

Qiskit 2.x improves parameter binding workflows.


````python
from qiskit.circuit import Parameter

theta = Parameter("θ")

qc_param = QuantumCircuit(1)
qc_param.rx(theta, 0)

qc_param
````

````python
bound_circuit = qc_param.assign_parameters({theta: 0.7})
bound_circuit
````

# 5. Observables and Operators

Qiskit 2.x standardizes observable representations using classes like:

- `SparsePauliOp`

These are widely used with **Estimator primitives**.


````python
from qiskit.quantum_info import SparsePauliOp

observable = SparsePauliOp.from_list([("ZZ",1)])

observable
````

### Example: Estimator Primitive

````python
from qiskit.primitives import StatevectorEstimator

estimator = StatevectorEstimator()

job = estimator.run([(qc, observable)])
job.result()
````

# 6. Why These Changes Matter

The Qiskit 2.x redesign reflects several long‑term goals:

### Stable execution interfaces
Primitives provide consistent APIs across simulators and hardware.

### Scalable experiments
Runtime services support:

- batching
- hybrid algorithms
- large circuit sets

### Hardware‑aware compilation
The transpiler increasingly incorporates hardware information.

### Modular ecosystem
Higher‑level tools (chemistry, ML, optimization) are separated into independent packages.
